```python
import torch
import gym
from crar.crar import CRARLightning
import numpy as np
```

```python
env = gym.make('SimpleMaze-v1')
print(env.id)
```

```python
state = env.reset()
```

## Making GIFs

```python
import PIL
from PIL import Image, ImageSequence
import numpy as np
state = env.reset()
ims = []
for _ in range(50):
    state = env.render()[0]
    env.step(np.random.randint(4))
    im = Image.fromarray(state * 256)
    im = im.resize((500, 500), resample=PIL.Image.NEAREST)
    ims.append(im)
ims[0].save('random_exploring.gif', format='GIF', append_images=ims[1:], save_all=True, duration=100, loop=0)
```

```python
import PIL
from PIL import Image, ImageSequence
import numpy as np
state = env.reset()
ims = []
goal_seq = [3, 3, 3, 3, 1, 1, 1]
for action in goal_seq:
    state = env.render()[0]
    env.step(action)
    im = Image.fromarray(state * 256)
    im = im.resize((500, 500), resample=PIL.Image.NEAREST)
    ims.append(im)
ims[0].save('to_goal.gif', format='GIF', append_images=ims[1:], save_all=True, duration=100, loop=0)
```

## Simple Plots

```python
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import host_subplot
import mpl_toolkits.axisartist as AA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
from matplotlib.patches import Circle, Rectangle
from matplotlib.offsetbox import AnchoredOffsetbox, TextArea, DrawingArea, HPacker
matplotlib.use("qt5agg")

%matplotlib inline
```

```python
# plt.figure()
def plot_maze(state):
    matplotlib.rcParams['figure.figsize'] = (10, 10)
    c = plt.pcolor(state[0], edgecolors='k', linewidths=1, cmap='RdBu', vmin=0.0, vmax=1.0)
    plt.show()
```
